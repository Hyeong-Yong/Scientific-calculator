```python
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

laser_wavelength = 1550* math.pow(10, -9) # [m]
crystal_length = 0.02 # [m]
EO_coefficients_33 = 31*math.pow(10, -12) #[m/V]
EO_coefficients_13 = 8.6*math.pow(10, -12) #[m/V]
n_e = 2.1381
n_o = 2.2112
E_field_pi = laser_wavelength /(crystal_length*(EO_coefficients_33 * math.pow(n_e, 3)-EO_coefficients_13*math.pow(n_o, 3)))

E_field_modulation = np.linspace(0, 800000, 100) 

print("E_pi :", E_field_pi)
I_0 =1
phase_0= 0
E_field= 1
I_out = (I_0/2) * (1 + np.sin(phase_0 + math.pi * E_field_modulation/E_field_pi))


plt.figure(figsize=(10,5))# 그래프를 표현할 액자 생성
plt.plot(E_field_modulation/1000, I_out)

plt.title('OVT sensor', fontsize = 15)
plt.xlabel('E-field [kV/m]')
plt.ylabel('Intensity [Normalized]', fontsize='large')
#current_values = plt.gca().get_yticks()
#plt.gca().set_xticklabels(['{:.0f}'.format(x) for x in current_values])
plt.ticklabel_format(axis='x',useOffset=False, style='plain')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.gca().xaxis.set_major_formatter(mticker.FormatStrFormatter("%.2f"))
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter("%.2f"))

plt.show() # 그래프 출력
#for i in range(len(E_field_modulation)):    
#    print (E_field_modulation[i] , end = '\n ')
    
    


for i in range(len(E_field_modulation)):    
    print (E_field_modulation[i] , end = '\n ')    

#for i in range(len(I_out)):    
#    print (I_out[i] , end = '\n ')
```
