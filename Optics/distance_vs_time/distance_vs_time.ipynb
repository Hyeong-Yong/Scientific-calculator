```python
import math
import matplotlib.pyplot as plt
import numpy as np


c = 3*math.pow(10, 8) # speed of light [m/s]
c_ps = c * math.pow(10, 3) / math.pow(10, 12) # [mm/ps]
ps = 1*math.pow(10, -12) # [ps]

t_ps = np.linspace(1, 10, 50) # 시간 [ps]
s_mm = c_ps*t_ps  # [mm/p]
plt.figure() # 그래프를 표현할 액자 생성
plt.plot(t_ps, s_mm)

plt.title('Line Plot : s = c * t', fontsize = 15)
plt.xlabel('Time [ps]')
plt.ylabel('Distance [mm]', fontsize='large')

plt.show() # 그래프 출력
```
