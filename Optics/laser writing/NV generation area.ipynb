```python
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import numpy as np

# Nat. Phontonics 11, 77 (2017)
# when diameter = 200nm, pulse energy = 25nJ

# 20 [nJ] : x [nJ] = 100^2 * pi [nm^2] : y^2 [nm^2]
# 1mJ은 되어야 100um의 면적을 갖는 NV 생성이 됨 

LaserPulseEnergy = np.linspace(20, 10**2, 1000)
Area = (100**2 * math.pi)*LaserPulseEnergy/2.5
Length= np.sqrt(Area)
plt.figure(figsize=(10,5))# 그래프를 표현할 액자 생성
plt.plot(LaserPulseEnergy, Length)

plt.title('NV generation Area', fontsize = 15)
plt.xlabel('Laser pulse energy [nJ] ')
plt.ylabel('Width [nm]', fontsize='large')
#current_values = plt.gca().get_yticks()
#plt.gca().set_xticklabels(['{:.0f}'.format(x) for x in current_values])
plt.ticklabel_format(axis='x',useOffset=False, style='plain')
plt.ticklabel_format(axis='y',useOffset=False, style='plain')
plt.gca().xaxis.set_major_formatter(mticker.FormatStrFormatter("%.2f"))
plt.gca().yaxis.set_major_formatter(mticker.FormatStrFormatter("%.2f"))

plt.show() # 그래프 출력

```
