```python
import math
import numpy as np
import matplotlib.pyplot as plt

G= 4.1667
Vref= 3
Din = np.array(range(0, 65535))
Vout = G * Vref *( 2* Din/65536-1)
plt.plot(Din, Vout, 'k-')  

plt.xlabel('Digits [number]')
plt.ylabel('Voltage [V]')
plt.yticks(np.arange(-12,13,2))
plt.title('Voltage vs Digits')
plt.grid(color= "gray", alpha=.5, linestyle = '-')
plt.legend()
plt.show()
Din = 0
Vout = G * Vref *( 2* Din/65536-1)
print('Min value: ', Vout)
Din = 65535
Vout = G * Vref *( 2* Din/65536-1)
print('Max value: ', Vout)
```
