```python
import math

thickness = 0.2# [mm]
refractive_index = 1.78 # PLA refractive index : 1.64 , craft paper : 1.78
c = 3*(10**8) # [m/s]

############################################8GHz########################################
frequency_a= 302* (10**9)# [Hz]
frequency_b= 310* (10**9)# [Hz]

lambda_a = c/frequency_a
lambda_b = c/frequency_b

lambda_a = lambda_a *1000 # m => mm
lambda_b = lambda_b *1000 

#print("lambda_a =", lambda_a ,"[mm]")
#print('lambda_b =', lambda_b ,'[mm]')

sythezised_lambda = (lambda_a*lambda_b)/ (abs(lambda_a-lambda_b))
print ('---------------8GHz-----------------')
print('sythezised_wavelength = ', sythezised_lambda, "[mm]")

phase_difference = 2*math.pi/sythezised_lambda * 2*thickness*(refractive_index-1) 
print('phase_difference = ', phase_difference, '[rad]')
print ('------------------------------------')




############################################9GHz########################################
frequency_a= 301.5* (10**9)# [Hz]
frequency_b= 310.5* (10**9)# [Hz]

c = 3*(10**8) # [m/s]

lambda_a = c/frequency_a
lambda_b = c/frequency_b

lambda_a = lambda_a *1000 # m => mm
lambda_b = lambda_b *1000 

#print("lambda_a =", lambda_a ,"[mm]")
#print('lambda_b =', lambda_b ,'[mm]')

sythezised_lambda = (lambda_a*lambda_b)/ (abs(lambda_a-lambda_b))
print ('---------------9GHz-----------------')
print('sythezised_wavelength = ', sythezised_lambda, "[mm]")

phase_difference = 2*math.pi/sythezised_lambda * 2*thickness*(refractive_index-1) 
print('phase_difference = ', phase_difference, '[rad]')
print ('------------------------------------')





############################################10GHz########################################

frequency_a= 301* (10**9)# [Hz]
frequency_b= 311* (10**9)# [Hz]

c = 3*(10**8) # [m/s]

lambda_a = c/frequency_a
lambda_b = c/frequency_b

lambda_a = lambda_a *1000 # m => mm
lambda_b = lambda_b *1000 

#print("lambda_a =", lambda_a ,"[mm]")
#print('lambda_b =', lambda_b ,'[mm]')

sythezised_lambda = (lambda_a*lambda_b)/ (abs(lambda_a-lambda_b))
print ('---------------10GHz-----------------')
print('sythezised_wavelength = ', sythezised_lambda, "[mm]")

phase_difference = 2*math.pi/sythezised_lambda * 2*thickness*(refractive_index-1) 
print('phase_difference = ', phase_difference, '[rad]')
print ('------------------------------------')





############################################11.5GHz########################################

frequency_a= 300.25* (10**9)# [Hz]
frequency_b= 311.75* (10**9)# [Hz]

c = 3*(10**8) # [m/s]

lambda_a = c/frequency_a
lambda_b = c/frequency_b

lambda_a = lambda_a *1000 # m => mm
lambda_b = lambda_b *1000 

#print("lambda_a =", lambda_a ,"[mm]")
#print('lambda_b =', lambda_b ,'[mm]')

sythezised_lambda = (lambda_a*lambda_b)/ (abs(lambda_a-lambda_b))
print ('---------------11.5GHz-----------------')
print('sythezised_wavelength = ', sythezised_lambda, "[mm]")

phase_difference = 2*math.pi/sythezised_lambda * 2*thickness*(refractive_index-1) 
print('phase_difference = ', phase_difference, '[rad]')
print ('------------------------------------')


############################################11.8GHz########################################

#frequency_a= 300.1* (10**9)# [Hz]
#frequency_b= 311.9* (10**9)# [Hz]

#c = 3*(10**8) # [m/s]

#lambda_a = c/frequency_a
#lambda_b = c/frequency_b

#lambda_a = lambda_a *1000 # m => mm
#lambda_b = lambda_b *1000 

#print("lambda_a =", lambda_a ,"[mm]")
#print('lambda_b =', lambda_b ,'[mm]')

#sythezised_lambda = (lambda_a*lambda_b)/ (abs(lambda_a-lambda_b))
#print ('---------------11.8GHz-----------------')
#print('sythezised_wavelength = ', sythezised_lambda, "[mm]")

#phase_difference = 2*math.pi/sythezised_lambda * 2*thickness*(refractive_index-1) 
#print('phase_difference = ', phase_difference, '[rad]')
#print ('------------------------------------')
```
